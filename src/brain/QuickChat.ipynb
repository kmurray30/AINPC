```python
# Reload the dependencies
%reload_ext autoreload
%autoreload 2

import sys

sys.path.insert(0, "../../..")
from src.core.Constants import Llm
from src.utils.ChatBot import ChatBot

# Call 4o to generate an initial prompt
chatbot = ChatBot(model=Llm.gpt_4o_mini)

precog_messages = []
precog_system_prompt = """
You are a character creator agent tasked with helping come up with prompts for creating characters.
"""
precog_messages.append({"role": "system", "content": precog_system_prompt})

responder_messages = []
responder_system_prompt = """
You are a writer tasked with creating a response from a given character to a given question
"""
responder_messages.append({"role": "system", "content": responder_system_prompt})

postcog_messages = []
postcog_system_prompt = """
You are a revisor tasked with scrutinizing the response generated by the writer and providing feedback on how to improve it.
"""
postcog_messages.append({"role": "system", "content": postcog_system_prompt})
```

```python
prompt = """
Given the following things about Raj, please generate a prompt that would instruct a "writer" agent into creating a reasonable response from Raj to the subsequent question.
Don't just repeat the information about Raj, but instead come up with a 1 or 2 angles to go for as a subset of the information provided.
List the angles in a bulleted list.

Raj info:
$$$RAJINFO$$$

Question:
$$$QUESTION$$$
"""

raj_info = """
* Objectifies women benignly. Respects them individually, but generally sees them as targets to impress or accessories to impress with.
* Tries to act cool and with swagger
* Acts chill, not too much energy if it's not called for
* Has a lack of physical fitness and a scrawny frame he wants to protect from overexertion. Acts kinda pre-madonna-y about it, but not in a feminine or flamboyant way.
* Pretty lazy
* Into swag and glam and pop culture, but not in a feminine way. In a wannabe way.
* Doesn't like "lame" things, but would jump on any opportunity to be a part of something where he could have a chance with women or have instagram potential
* Overall a decent guy who doesn't cross the line into innappropriate behavior, but is definitely a little clueless and self-centered.
"""

question = "Want to come to yoga with me later?"

# Call 4o-mini to generate the precog prompt
prompt = prompt.replace("$$$QUESTION$$$", question)
prompt = prompt.replace("$$$RAJINFO$$$", raj_info)
precog_messages.append({"role": "user", "content": prompt})
precog_response = chatbot.call_chat_agent(precog_messages)
precog_messages.pop()
print("Precog: " + precog_response)

precog_response += """
Keep your response short and to the point. Max 1-2 sentences.
Don't use cheesey catch-phrases or cliches
"""

# Call 4o-mini to generate the response
responder_messages.append({"role": "user", "content": precog_response})
responder_messages.append({"role": "user", "content": question})
responder_response = chatbot.call_chat_agent(responder_messages)
responder_messages.pop()
responder_messages.pop()
print("\nResponder: " + responder_response)

# postcog_prompt = f"""
# Given the following "response" from a writer attempting to create a response from Raj to the question, please provide feedback on how to improve it.
# Address the following points:
# 1. Was anything in the response cheesey or cliche?
# 2. Did the response align with Raj's character traits and personality?
# 3. If metaphors or references were used, did they actually make sense? Were they logically consistent or narratively appropriate? Use scrutiny here.

# Raj info:
# {raj_info}

# Question:
# {question}

# Response:
# {responder_response}
# """

# # Call 4o-mini to generate the postcog response
# postcog_messages.append({"role": "user", "content": postcog_prompt})
# postcog_response = chatbot.call_chat_agent(postcog_messages)
# postcog_messages.pop()
# print("\nPostcog: " + postcog_response)
```
